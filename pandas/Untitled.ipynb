```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read the csv file 
olympics_data = pd.read_csv('../data/olympics.csv',skiprows=4)
print(olympics_data)
```

```python
#get the first olympics
first = olympics_data[olympics_data.Edition == 1896]
```

```python
first.Sport.value_counts()
```

```python
first.Sport.value_counts().plot(kind='line', figsize=(10,3));
```

```python
first.Sport.value_counts().plot(kind='bar');
```

```python
first.Sport.value_counts().plot(kind='barh');
```

```python
first.Sport.value_counts().plot(kind='pie');
```
